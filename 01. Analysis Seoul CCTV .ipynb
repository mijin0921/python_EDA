### pandas에서 데이터 프레임을 병합하는 방법
- pd.concat()
- pd.merge()
- pd.join()

```python
#딕셔너리 안에 리스트 형태
#리스트가 열로 추가됨

left = pd.DataFrame({ 
    'Key' : ['K0','K4','K2','K3'],
    'A' : ['A0','A1','A2','A3'],
    'B' : ['B0','B1','B2','B3']
})
left
```

```python
#리스트 안의 딕셔너리 형태
#딕셔너리가 행으로 추가됨

right = pd.DataFrame([
     {'Key': 'K0' ,'C':'C0','D':'D0'},
     {'Key': 'K1' ,'C':'C1','D':'D1'},
     {'Key': 'K2' ,'C':'C2','D':'D2'},
     {'Key': 'K3' ,'C':'C3','D':'D3'}
])
right
```

## pd.merge
- 두 데이터 프레임에서 컬럼이나 인덱스를 기준으로 잡고 데이터를 병합하는 방법
- 기준이 되는 컬럼이나 인덱스를 키값이라고 함
- 기준이 되는 키 값은 두 데이터 프레임에 모두 포함되어 있어야 한다

```python
pd.merge(left, right,on='Key')
#how='inner'가 기본으로 세팅되어 있음
```

```python
pd.merge(left, right,how='left',on='Key')
```

```python
pd.merge(left, right,how='right',on='Key')
```

```python
pd.merge(left, right,how='outer',on='Key')

#이렇게 했을떄 NaN값을 어떻게 처리할지 고민해야함
```

### 예제1 : 그래프 기초

### 삼각함수 그리기 
- np.arange(a,b,s) :a부터 b까지 S간격
- np.sin(value)

```python
t = np.arange(0,12,0.01)

plt.figure(figsize=(10,6))
plt.plot(t,np.sin(t))
plt.plot(t,np.cos(t))
plt.show()
```

- 1. 격자무늬 추가
- 2. 그래프 제목 추가
- 3. x축y축 제목 추가
- 4. 주황색 파란색 선 데이터 의미 구분

```python
def drawGragh():


    plt.figure(figsize=(10,6))
    plt.plot(t,np.sin(t))
    plt.plot(t,np.cos(t))
    plt.grid(True)
    plt.title("Example of sinwave")
    plt.xlabel("Time")
    plt.ylabel('Amplittude') #진폭
    plt.legend(labels=['sin','cos'], loc='upper right') #doc 스트링 참조하면서 하면됨
    plt.show()
```

```python
drawGragh()
```

### numpy를 이용한 선형회귀선 만들기 
- np.polyfit() : 직선을 구성하기 위한 계수를 계산
- np.poly1d() : polyfit으로 찾은 계수로 파이썬에서 활용할 수 있는 함수로 만들어 준다

```python
import numpy as np
```

```python
fd1 = np.polyfit(dataResult['인구수'],dataResult['소계'],1)
fd1
```

```python
f1 = np.poly1d(fd1)
f1
```

```python
f1(400000) #인구가 40만인 구에 설치된 CCTV 개수 예측
```

```python
fx = np.linspace(100000,700000,100) 
```

- 경향선을 그리기 위한 x 데이터 생성
- np.linspace(a,b,n) : a부터 b까지 n개의 등간격 데이터 생성

```python
plt.figure(figsize=(14,10))
plt.scatter(dataResult['인구수'],dataResult['소계'],s =50)
plt.plot(fx,f1(fx),color='g',ls='--',lw=3)
plt.grid(True)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.show()
```

## 7. 강조하고 싶은 데이터를 시각화해보자

### 그래프 다듬기
#### 경향과의 오차 만들기
- 경향(trend)과의 오차 만들기
- 경향은 f1 함수에 해당 인구를 입력
- f1(dataResult['인구수'])

```python
dataResult['오차'] = dataResult['소계']-f1(dataResult['인구수'])
dataResult.head()
```

```python
#경향과 비교해서 데이터의 오차가 너무 나는 데이터를 계산

df_sort_f = dataResult.sort_values(by='오차',ascending=False) #내림차순
df_sort_t = dataResult.sort_values(by='오차',ascending=True) #오름차순
```

```python
#경향대비 CCTV를 많이가진 구
df_sort_f.head()
```

```python
#경향대비 CCTV 적게 가진 구
df_sort_t.head()
```

```python
from matplotlib.colors import ListedColormap

#colormap 을 사용자 정의로 세팅
color_step=['#e74c3c','#2ecc71','#95a9a6','#2ecc71','#3498db','#3498db']
my_cmap=ListedColormap(color_step)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(dataResult['인구수'],dataResult['소계'],s =50,c=dataResult['오차'], cmap=my_cmap)
plt.plot(fx,f1(fx),color='g',ls='--',lw=3)
for i in range(5):
    
    #상위 5개
    plt.text(df_sort_f['인구수'][i]*1.02,df_sort_f['소계'][i]*0.98,df_sort_f.index[i])

    #하위5개
    plt.text(df_sort_t['인구수'][i]*1.02,df_sort_t['소계'][i]*0.98,df_sort_t.index[i])
plt.colorbar()
plt.grid(True)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.show()
```
