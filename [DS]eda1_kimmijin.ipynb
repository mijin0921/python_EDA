## 문제 3)

```python
ediya_df = pd.read_csv('../data/eda1_ediya.csv',encoding='utf-8',usecols={'구','매장이름','주소'})
ediya_df.head()
```

```python
starbucks_df = pd.read_csv('../data/eda1_starbucks.csv',encoding='utf-8',usecols={'구','매장이름','주소'})
starbucks_df.head()
```

```python
ediya_num = ediya_df.pivot_table(index='구',aggfunc=np.count_nonzero)
ediya_num
```

```python
ediya_num.rename(columns={'매장이름' : '매장수'},inplace=True)
ediya_num.drop(columns='주소',inplace=True)
ediya_num
```

```python
starbucks_num = starbucks_df.pivot_table(index='구',aggfunc=np.count_nonzero)
starbucks_num.rename(columns={'매장이름' : '매장수'},inplace=True)
starbucks_num.drop(columns='주소',inplace=True)
```

```python
cafes_num=pd.merge(left=starbucks_num, right=ediya_num, on='구')
```

```python
cafes_num
```

```python
cafes_num.rename(columns={ '매장수_x' : '스타벅스', '매장수_y' : '이디야'},inplace=True)
```

```python
cafes_num
```

```python
cafes_num.corr()
```

### 분석내용
- 스타벅스 매장수와 이디야 매장수의 상관관계 분석 차트로 보면 두 브랜드 매장수의 상관계수는 0.63 정도로 높은 관계가 있다. 강서구, 용산구,강북구를
제외하면 스타벅스 매장수가 늘어남에 따라 이디야 매장수도 늘어나고 있다.

- 시각화한 지도를 보면 스타벅스 매장 위치와 이디야 매장위치를 원으로 표현하고 원의 넓이로 반경 100m를 표현했다.(반경 100m정도가 서로 가깝다고 가정)
그리고 구별 이디야 매장수를 색깔로 비교했다.  

-전반적으로 이디야 매장수는 구별로 고르게 입점해 있지만 스타벅스는 특정 구에 집중되어 있는것을 알 수 있다.

-이디야 매장수가 상대적으로 많은 구에(구 면적의 색깔이 진한 곳) 스타벅스 매장이 집중되어 있는 것을 보아, 이 구는 카페수가 많은 번화가일 것으로 추측된다.

-스타벅스 매장이 집중되어 있는 곳에 이디야 매장의 반경넓이가 겹치는 곳이 많이 보인다.

-따라서 이디야 매장은 서울시 전체적으로 고르게 입점되어 있지만, 스타벅스 매장이 상대적으로 집중되어 있고, 번화가일 것으로 추측되는 구에서
스타벅스 주변에 많이 분포되어 있는 것으로 해석할 수 있다.
