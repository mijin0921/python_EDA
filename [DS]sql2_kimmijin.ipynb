위도, 경도 정보를 이용하여 미왕빌딩에서 1킬로 이내에 위치한 주유소 정보를 검색하세요.

- 주유소 아이디, 주유소 브랜드명, 주유소 매장명, 주소, 미왕빌딩으로 부터의 거리 (km)

제출 6.
- 쿼리, 결과

```python
# 미왕빌딩 위도와 경도 및 거리 제한 설정
cur.execute("SET @miwang_lat = 37.564213")
cur.execute("SET @miwang_lng = 126.974179")
cur.execute("SET @distance_limit = 1")

# 주유소 정보 검색(하버사인 공식)
sql_5 = """
    SELECT 
        s.id AS station_id,
        b.name AS brand_name,
        s.name AS station_name,
        s.address,
        (6371 * ACOS(
        COS(RADIANS(@miwang_lat)) * COS(RADIANS(s.lat)) * COS(RADIANS(s.lng) - RADIANS(@miwang_lng)) + 
        SIN(RADIANS(@miwang_lat)) * SIN(RADIANS(s.lat))
        )) AS distance
    FROM 
        GAS_STATION s
    INNER JOIN 
        GAS_BRAND b ON s.brand = b.id
    HAVING 
        distance <= @distance_limit
    ORDER BY 
        distance ASC
"""

cur.execute(sql_5)
result = cur.fetchall()

# 컬럼명 지정
columns = ["station_id", "brand_name", "station_name", "address", "distance"]

# 데이터 프레임 생성
df = pd.DataFrame(result, columns=columns)
df
```

위도, 경도 정보를 이용하여 미왕빌딩에서 셀프주유가 가능하고 24시간이면서 편의점이 있는 가장 가까운 주유소 10개를 휘 발유 가격이 가장 저렴한 순으로 정렬하여 조회하세요.

- 주유소 아이디, 주유소 브랜드명, 주유소 매장명, 주소, 휘발유 가격, 부가정보 (셀프, 24시간, 편의점 여부), 미왕빌딩으로 부터의 거리 (km)


제출 7.
- 쿼리, 결과

```python
sql_6= """SELECT
    s.id AS station_id,
    b.name AS brand_name,
    s.name AS station_name,
    s.address,
    s.gasoline,
    s.self, 
    s.24_hours,
    s.convenience_store,
    (6371 * ACOS(
        COS(RADIANS(@miwang_lat)) * COS(RADIANS(s.lat)) * COS(RADIANS(s.lng) - RADIANS(@miwang_lng)) +
        SIN(RADIANS(@miwang_lat)) * SIN(RADIANS(s.lat))
    )) AS distance
FROM 
    GAS_STATION s
INNER JOIN 
    GAS_BRAND b ON s.brand = b.id
WHERE 
    s.self = 1 
    AND s.24_hours = 1 
    AND s.convenience_store = 1
ORDER BY 
    s.gasoline ASC,
    distance ASC
LIMIT 10"""

cur.execute("SET @miwang_lat = 37.564213")
cur.execute("SET @miwang_lng = 126.974179")
cur.execute(sql_6)
result = cur.fetchall()
columns = [desc[0] for desc in cur.description]
df = pd.DataFrame(result, columns=columns)
df
```

제출 8.
- 쿼리, 결과

```python
sql_7 = """
SElECT
s.gu,
b.name,
AVG(s.gasoline) AS Oil_price
FROM GAS_STATION s
INNER JOIN GAS_BRAND b ON s.brand = b.id
GROUP BY s.gu, b.name
ORDER BY Oil_price"""

cur.execute(sql_7)
result = cur.fetchall()
columns = [desc[0] for desc in cur.description]
df = pd.DataFrame(result, columns=columns)
df
```
