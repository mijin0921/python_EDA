```python
def changeIng(text):
    # ì¡°ê±´ 1
    con1 = r'[.]$'
    parts = text.split(',')  # ë¬¸ìì—´ì„ ì½¤ë§ˆë¡œ ë¶„ë¦¬

    for i in range(len(parts)):
        # parts[i] = parts[i].strip()  # ê° ë¶€ë¶„ì—ì„œ ì•ë’¤ ê³µë°± ì œê±°
        match1 = re.search(con1, parts[i])
        if match1:
            parts[i] = parts[i][:-1]  # ê° ë¶€ë¶„ì—ì„œ ë§ˆì§€ë§‰ ë§ˆì¹¨í‘œ ì œê±°

    # ëª¨ë“  ë¶€ë¶„ì„ ë‹¤ì‹œ ê²°í•©
    result = ', '.join(parts)

    # ì¡°ê±´ 2
    con2 = r'[.] May Contain.'
    match2 = re.search(con2, result)
    if match2:
        n = match2.start()
        result = result[:n]

    return result

# DataFrameì˜ 'Ingredients' ì—´ì— ëŒ€í•´ í•¨ìˆ˜ ì ìš©
tmp = [changeIng(i) for i in df_target['Ingredients']]
df_target['Ingredients'] = tmp

print(df_target['Ingredients'])
```
