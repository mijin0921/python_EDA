```python
exchangeList=soup.select('#exchangeList > li')
len(exchangeList),exchangeList
```

```python
import pandas as pd

exchange_data = []
baseurl = 'http://finance.naver.com'

for item in exchangeList:
    data = {'title' : item.select_one('.h_lst').text,
            'exchange' : item.select_one('span.value').text,
            'change': item.select_one('span.change').text,
            'updown' : item.find('div','head_info').css.filter('span')[3].text,
            'link' : baseurl + item.select_one('a').get('href')}
    print(data)
    exchange_data.append(data)
            
            
```

```python
df = pd.DataFrame(exchange_data)
```
