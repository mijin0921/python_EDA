- 이디야 매장 주요 분포 지역 (매장수가 많은 상위 5개 구이름, 매장 개수 출력)
- 구별 브랜드 각각의 매장 개수 조회 (구이름, 브랜드이름, 매장 개수 출력)
- 구별 브랜드 각각의 매장 개수 조회 (구이름, 스타벅스 매장 개수, 이디야 매장 개수 출력)

```python
sql_1 = 'select gu_name , count(name) 매장수 from coffee_store where brand = 1 group by gu_name order by 매장수 desc limit 5'
sql_2 = 'select gu_name , count(name) 매장수 from coffee_store where brand = 2 group by gu_name order by 매장수 desc limit 5'
sql_3 = 'select gu_name , brand, count(name) 매장수 from coffee_store group by gu_name , brand'
sql_4 = """SELECT e.gu_name, COUNT(e.name) AS 스타벅스매장수, s.이디야매장수 AS 이디야매장수 FROM coffee_store e 
inner JOIN (SELECT gu_name, COUNT(name) AS 이디야매장수 FROM coffee_store WHERE brand = 2 GROUP BY gu_name) s 
ON e.gu_name = s.gu_name 
WHERE e.brand = 1 GROUP BY e.gu_name"""
```

```python
cur.execute(sql_1)
cur.execute(sql_2)
cur.execute(sql_3)
cur.execute(sql_4)
```

시각화 프로젝트를 위하여 다음의 규칙으로 쿼리하여 CSV 파일로 저장합니다. (Python 코드로 작업)

- 전체 데이터를 가져오는데, 각 스타벅스 매장별로 이디야 전체 매장정보가 매칭되어 있어야 합니다. (정렬 : s_id, e_id 순)
- 다음의 형식으로 저장되어야 합니다. (브랜드 이름, 칼럼 명 주의)
- 데이터 프레임 출력을 해주세요. 데이터 프레임 미출력시 감점입니다.

```python
sql_all = """SELECT s.id AS s_id,
(SELECT name FROM coffee_brand WHERE id = 1) AS s_brand, 
s.name AS s_name, s.gu_name AS s_gu, s.address AS s_address,
s.lat AS s_lat,s.lng AS s_lng, 
e.id AS e_id, (SELECT name FROM coffee_brand WHERE id = 2) AS e_brand,
e.name AS e_name,
e.gu_name AS e_gu,
e.address AS e_address,
e.lat AS e_lat,
e.lng AS e_lng
FROM coffee_store s
LEFT JOIN coffee_store e ON s.gu_name = e.gu_name"""
```

```python
cur.execute(sql_all)
result = cur.fetchall()
```

```python
import pandas as pd
df = pd.DataFrame(result)
df.head()
```
